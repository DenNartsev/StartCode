```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
f = open('forestfires.csv','r')
data = pd.read_csv('forestfires.csv')
data
```

```python
def get_data(df):
    # month_and_days_in_number
    month_w = ['jan', 'feb', 'dec', 'mar', 'apr', 'oct', 'nov', 'may']
    #month_sp = ['jun', 'apr', 'may']
    month_sm = ['sep', 'jul', 'aug', 'jun']
    #month_au = ['jan', 'oct', 'nov']
    #set_per = [month_w, month_sm]
    
    for month in month_sm:
        df['month'][df.month == month] = 1
    for month in month_w:
        df['month'][df.month == month] = 0 
    days_work = ['mon', 'tue', 'wed', 'thu', 'fri']
    days_hol = ['sat', 'sun']
    for day in days_hol:
        df['day'][df.day == day] = 1 
    for day in days_work:
        df['day'][df.day == day] = 0
    return df
```

```python
X_data = data.iloc[:,0:-1]
X_data = get_data(X_data)
X_data = X_data.rename(columns={'month' : 'per'})
#X_data = pd.get_dummies(X_data, columns=['per'])
X_data = X_data.drop(columns=['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH'])
X_data
```

```python
X_data = X_data.astype('float64')
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
X_data_s = scale(X_data)
Y_data = np.array(data['area'], dtype='float64')
Y_data = np.log(1 + Y_data)

X_data_train, X_data_test, Y_data_train, Y_data_test = train_test_split(X_data, Y_data, test_size=0.4)
X_data_train_s, X_data_test_s, Y_data_train_s, Y_data_test_s = train_test_split(X_data, Y_data, test_size=0.4)
```

```python
from sklearn import linear_model
reg_s = linear_model.Ridge(alpha = .5)
reg_s.fit(X_data_train_s,Y_data_train_s) 
w1 = reg_s.coef_
reg = linear_model.Ridge(alpha = .5)
reg.fit(X_data_train,Y_data_train) 
w2 = reg.coef_
```

## With standardization

## Train

```python
x = np.dot(X_data_train_s,w1)
y = Y_data_train_s
p1 = np.poly1d(np.polyfit(x,y,1))
plt.scatter(x,y,color='r')
plt.plot(x,p1(x),'g')
plt.show()
```

## Test

```python
x = np.dot(X_data_test_s,w1)
y = Y_data_test_s
p1 = np.poly1d(np.polyfit(x,y,1))
plt.scatter(x,y,color='r')
plt.plot(x,p1(x),'g')
plt.show()
```

```python
plt.plot(X_data.columns,w1)
plt.show()
```

### Error

```python
class_test_size = np.arange(.05, 1, .05)
error_test = np.array([])
error_train = np.array([])
for size in class_test_size:
    k_train = len(data) * (1 - size)
    k_test = len(data) * size
    Q_train = 0
    Q_test = 0
    for i in range(1000):
        X_data_train_s, X_data_test_s, Y_data_train_s, Y_data_test_s = train_test_split(X_data, Y_data, test_size=size)
        reg_s.fit(X_data_train_s,Y_data_train_s) 
        w1 = reg_s.coef_
        Q_train += np.linalg.norm(Y_data_train_s - np.dot(X_data_train_s, w1), ord=2)
        Q_test += np.linalg.norm(Y_data_test_s - np.dot(X_data_test_s, w1), ord=2)
    if size == 0.4:
        print('Error Train   Error Test')
        print((Q_train / k_train).round(2), (Q_test / k_test).round(2), sep='\t\t')
    error_train = np.append(error_train, Q_train / 1000 / k_train)
    error_test = np.append(error_test, Q_test / 1000 / k_test)
```

```python
plt.plot((1 - class_test_size)[::-1], error_train[::-1])
plt.title("Train")
plt.ylabel('Error')
plt.xlabel('Train_size')
plt.show()
```

```python
plt.plot(class_test_size, error_test)
plt.title("Test")
plt.ylabel('Error')
plt.xlabel('Test_size')
plt.show()
```
