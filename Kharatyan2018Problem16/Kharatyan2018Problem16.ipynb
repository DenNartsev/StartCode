```python
#Функция для нахождения евклидовой метрики
def dist(a,b):
    c = 0
    for i in range(9):
        c += (a[i]-b[i])**2
    return np.sqrt(c)
```

### Вводим ядерные функции

```python
def K_1(r):
    assert (abs(r)<=1),"Bad argument!"
    return 3/4*(1-r**2)

def K_2(r):
    assert (abs(r)<=1),"Bad argument!"
    return 15/16*((1-r**2)**2)

def K_3(r):
    assert (abs(r)<=1),"Bad argument!"
    return 1-abs(r)

def K_4(r):
    return ((2*np.pi)**(-0.5))*(np.exp(-0.5*r**2))

def K_5(r):
    assert (abs(r)<=1),"Bad argument!"
    return 0.5*abs(r)
```

```python
def count_errors(h):
    errors = 0
    for i in range(100):
        lov_num = 0
        class1_range = 0
        class2_range = 0
        for j in range(100):
            if i != j and data["Diagnosis"][j] == 1:
                class1_range += K_1(metrics[i][j]/h) * data["Diagnosis"][j]
                #print(data["Diagnosis"][j])
                #print(K_4(metrics[i][j]/h) * data["Diagnosis"][j])
            elif i != j and data["Diagnosis"][j] == -1:
                class2_range += K_1(metrics[i][j]/h) * data["Diagnosis"][j]
            else:
                pass
        if data["Diagnosis"][i] == 1:
                class1_range = class1_range/87
                class2_range = class2_range/12
        if data["Diagnosis"][i] == -1:
            class1_range = class1_range/88
            class2_range = class2_range/11
        lov_num = class1_range + class2_range
        #print(i)
        #print("LOV_NUM:",lov_num)
        #print(class1_range)
        #print(class2_range)
        if np.sign(data["Diagnosis"][i]) != np.sign(lov_num):
            errors += 1
        #print("Diagnosis:",data["Diagnosis"][i])
        #print("Calculated:",np.sign(lov_num))
        #print(np.sign(data["Diagnosis"][i]) != np.sign(lov_num))
    return errors
```
