```python
#источник: "Introduction to Machine Learning with Python: A Guide for Data Scientists"
#           by Andreas C. Muller & Sarah Guido


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=33)
```

```python
training_accuracy = []
test_accuracy = []
neighbors_arr = range(1, 11)
for k in neighbors_arr:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_arr, training_accuracy, label="accuracy on a training set")
plt.plot(neighbors_arr, test_accuracy, label="accuracy on a test set")
plt.ylabel("accuracy")
plt.xlabel("k")
plt.legend()

```
