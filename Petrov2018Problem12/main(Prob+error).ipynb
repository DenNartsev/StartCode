```python
dataset=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
dataset.columns = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD/OD of diluted wines', 'Proline ']
dataset[:15]
```

### График

### Стандартная ошибка для евклид

```python
k = 16
test_size = range(10, 100, 10)

means = {}
for size in test_size:
    X_train,X_test,y_train,y_test=sklearn.model_selection.train_test_split(wine_data, wine_class, test_size=size/100, random_state=42)
    classifier = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    classifier.fit(X_train, y_train)
    prediction = classifier.predict(X_test)
    
    error = 1 - scp.mean(prediction == y_test)
    std = scp.std(prediction == y_test)
    means[size/100] = [error, std]
```

```python
sizes = list(means.keys())
errors = []
stds = []
for item in means.values():
    errors.append(item[0])
    stds.append(item[1])
    
fig, ax = plt.subplots()
ax.plot(sizes, errors, color="blue", label="errors")
ax.plot(sizes, stds, color="green", label="standard errors")
ax.set_xlabel("Sizes")
ax.legend()
plt.show()
```
