# Семантическое Сходство Текста

Цель работы: Учитывая два предложения текста, s1 и s2, системы должны вычислить, насколько похожи s1 и s2, возвращая оценку сходства. Аннотации и системы будут использовать шкалу от 0 (Без отношения) до 5 (семантическая эквивалентность), указывая на сходство между двумя предложениями. Если оценка сходства выше 2,5 возвращать значение 1 ("похожи"), иначе возвращать значение 0("не похожи").

Описание выборки предоставлено ниже через графики распределения слов и количества символов.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, HashingVectorizer
from sklearn import pipeline, preprocessing, feature_extraction


from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
from sklearn.utils.fixes import signature
from sklearn.metrics import average_precision_score
from sklearn.metrics import log_loss, accuracy_score

import scipy
```

```python
df = pd.read_csv('sts2015-en-post/data/clean/text.clean', sep=' ', delimiter='\t', header=None)
df = df.sample(frac=1)
df.index=range(len(df))
df.columns = ['y', '1', '2', '3', 'q1', 'q2']
df = df[['y', 'q1', 'q2']]
df['y'] = df['y'].apply(lambda x: x > 2.5)
```

#### Основной подход

```python
df['q1'] = df['q1'].apply(lambda x: str(x).lower())
df['q2'] = df['q2'].apply(lambda x: str(x).lower())
```

```python
from gensim.models import KeyedVectors

# Creating the model
en_model = KeyedVectors.load_word2vec_format('wiki.simple.vec')

# Getting the tokens 
words = []
for word in tqdm(en_model.vocab):
    words.append(word)
```

```python
def get_emb(x):
    tmp = []
    for word in x:
        try:
            tmp += [en_model[word]]
        except:
            None
    tmp = np.array(tmp)
    if len(tmp) == 0:
        return np.zeros(300)
    else:
        return tmp.mean(axis=0)
```

Извлечем центроиды предложений

```python
q1_cetr = np.array(df['q1'].apply(lambda x: get_emb(str(x).split(' '))).values.tolist())
q2_cetr = np.array(df['q2'].apply(lambda x: get_emb(str(x).split(' '))).values.tolist())
```

### Посчитаем различные расстояния между центроидами

```python
d1 = np.expand_dims(list(map(scipy.spatial.distance.cosine, q1_cetr, q2_cetr)), axis=-1)
d2 = np.expand_dims(list(map(scipy.spatial.distance.canberra, q1_cetr, q2_cetr)), axis=-1)
d3 = np.expand_dims(list(map(scipy.spatial.distance.cityblock, q1_cetr, q2_cetr)), axis=-1)
d4 = np.expand_dims(list(map(scipy.spatial.distance.euclidean, q1_cetr, q2_cetr)), axis=-1)
```

### Сформируем итоговую выборку
Она состоит из центроид предложений, модуля разности центроид, а так же различных расстояний

```python
X = pd.DataFrame(np.concatenate([q1_cetr, q2_cetr, np.abs(q1_cetr - q2_cetr), d1,d2,d3,d4], axis=1))
```

## Линейная регрессия
Структрные параметры - l1 l2 регуляризация

Дополнительно будем считать logloss

```python
from sklearn.linear_model import LogisticRegression
import lightgbm as lgbm 
#model = lgbm.LGBMClassifier()
mapa={}


ap_arr = []
for fold in range(5):
    train_size =1.
    model = LogisticRegression(penalty='l1', random_state=0)
    curr_len = X[X.fold != fold].drop(['y', 'fold'], axis=1).shape[0]
    train_x = X[X.fold != fold].drop(['y', 'fold'], axis=1)[:int(curr_len*train_size)]
    train_y = X[X.fold != fold]['y'][:int(curr_len*train_size)]

    model.fit(train_x, train_y)
    pred = model.predict_proba(X[X.fold == fold].drop(['y', 'fold'], axis=1))[:, 1]

    precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
    average_precision = average_precision_score(X[X.fold == fold]['y'], pred)
    ap_arr += [average_precision]
ap_arr = np.array(ap_arr)    



print('Mean average_precision: {}, std: {}'.format(ap_arr.mean(), ap_arr.std()))

precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
average_precision = average_precision_score(X[X.fold == fold]['y'], pred)

step_kwargs = ({'step': 'post'}
               if 'step' in signature(plt.fill_between).parameters
               else {})
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.4f}, logloss={1:0.4f}'.format(
          average_precision, log_loss(X[X.fold == fold]['y'], pred)))
plt.show()
```

## Knn
Для него построим график качества от k со стандартным отклонением

```python
from sklearn.neighbors import KNeighborsClassifier



arr = []
arr_std = []
for k in tqdm([1,3,5,10,15]):
    ap_arr = []
    for fold in range(5):
        train_size = 1.
        model = KNeighborsClassifier(n_neighbors=k)

        curr_len = X[X.fold != fold].drop(['y', 'fold'], axis=1).shape[0]
        train_x = X[X.fold != fold].drop(['y', 'fold'], axis=1)[:int(curr_len*train_size)]
        train_y = X[X.fold != fold]['y'][:int(curr_len*train_size)]

        model.fit(train_x, train_y)
        pred = model.predict_proba(X[X.fold == fold].drop(['y', 'fold'], axis=1))[:, 1]

        precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
        average_precision = average_precision_score(X[X.fold == fold]['y'], pred)
        ap_arr += [average_precision]
    ap_arr = np.array(ap_arr)    

    arr += [ap_arr.mean()]
    arr_std += [ap_arr.std()]
    


    precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
    average_precision = average_precision_score(X[X.fold == fold]['y'], pred)

    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(plt.fill_between).parameters
                   else {})
    plt.step(recall, precision, color='b', alpha=0.2,
             where='post')
    plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('2-class Precision-Recall curve: AP={0:0.4f}, logloss={1:0.4f}'.format(
              average_precision, log_loss(X[X.fold == fold]['y'], pred)))
    plt.show()
```

```python
plt.fill_between([1,3,5,10,15], np.array(arr) - np.array(arr_std), np.array(arr) + np.array(arr_std))
```

## Дерево решений
Для него будем строить график зависимости качества от глубины(для 'gini' и 'entropy' отдельно)

```python
from sklearn import tree

mapa = {}
for criterion in ['gini', 'entropy']:

    arr = []
    arr_std = []
    for max_depth in tqdm([1,3,5,7,9,10]):
        ap_arr = []
        for fold in range(5):
            train_size = 1.
            model = tree.DecisionTreeClassifier(max_depth=max_depth, criterion=criterion)

            curr_len = X[X.fold != fold].drop(['y', 'fold'], axis=1).shape[0]
            train_x = X[X.fold != fold].drop(['y', 'fold'], axis=1)[:int(curr_len*train_size)]
            train_y = X[X.fold != fold]['y'][:int(curr_len*train_size)]

            model.fit(train_x, train_y)
            pred = model.predict_proba(X[X.fold == fold].drop(['y', 'fold'], axis=1))[:, 1]

            precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
            average_precision = average_precision_score(X[X.fold == fold]['y'], pred)
            ap_arr += [average_precision]
        ap_arr = np.array(ap_arr)    

        arr += [ap_arr.mean()]
        arr_std += [ap_arr.std()]

    mapa[criterion] = (arr, arr_std)

    precision, recall, _ = precision_recall_curve(X[X.fold == fold]['y'], pred)
    average_precision = average_precision_score(X[X.fold == fold]['y'], pred)

    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(plt.fill_between).parameters
                   else {})
    plt.step(recall, precision, color='b', alpha=0.2,
             where='post')
    plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('2-class Precision-Recall curve: AP={0:0.4f}, logloss={1:0.4f}'.format(
              average_precision, log_loss(X[X.fold == fold]['y'], pred)))
    plt.show()
```

```python
plt.fill_between([1,3,5,7,9,10], np.array(mapa['gini'][0]) - np.array(mapa['gini'][1]), np.array(mapa['gini'][0]) + np.array(mapa['gini'][1]))
```

```python
plt.fill_between([1,3,5,7,9,10], np.array(mapa['entropy'][0]) - np.array(mapa['entropy'][1]), np.array(mapa['entropy'][0]) + np.array(mapa['gini'][1]))
```
