```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
f = open('forestfires.csv','r')
data = pd.read_csv('forestfires.csv')
data
```

```python
data.info()
```

```python
def get_data(df):
    # month_and_days_in_number
    names_month = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
    for j in range(12):
        df['month'][df.month == names_month[j]] = j + 1 
    names_days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
    for j  in range(7):
        df['day'][df.day == names_days[j]] = j + 1       
    return df
```

```python
X_data = data.iloc[:,0:-1]
X_data = get_data(X_data)
X_data = X_data.astype('float64')
from sklearn.preprocessing  import scale
X_data_s = scale(X_data)
Y_data = np.array(data['area'], dtype='float64')
Y_data = np.log(1 + Y_data)
```

```python
from sklearn import linear_model
reg_s = linear_model.Ridge(alpha = .5)
reg_s.fit(X_data_s,Y_data) 
w1 = reg_s.coef_
reg = linear_model.Ridge(alpha = .5)
reg.fit(X_data,Y_data) 
w2 = reg.coef_
```

## With standardization

```python
x = np.dot(X_data_s,w1)
y = Y_data
p1 = np.poly1d(np.polyfit(x,y,1))
plt.scatter(x,y,color='r')
plt.plot(x,p1(x),'g')
plt.xlim(-1,1.5)
plt.show()
```

```python
plt.plot(np.arange(1, 13), np.absolute(w1))
plt.show()
```

### Error

```python
Q = np.linalg.norm(Y_data - np.dot(X_data_s, w1), ord=2)
Q
```

## Without standardization

```python
x = np.dot(X_data,w2)
y = Y_data
p1 = np.poly1d(np.polyfit(x,y,1))
plt.scatter(x,y,color='r')
plt.plot(x,p1(x),'g')
plt.xlim(-0.5, 3)
plt.show()
```

```python
plt.plot(np.arange(1, 13), np.absolute(w2))
plt.show()
```
